```python
# drivers and active years
_, _, counts = plt.hist(df.Years_Active, bins = range(df.Years_Active.min(), df.Years_Active.max() + 2), align = 'left', rwidth = 0.9)
plt.xticks(range(df.Years_Active.min(), df.Years_Active.max() + 1))
plt.xlabel('Active Years')
plt.ylabel('Drivers')
plt.bar_label(counts)
plt.show()
```

```python
# explore Experience and Driver's Performance
c1 = df.Race_Starts.corr(df.Championships)
c2 = df.Race_Starts.corr(df.Race_Wins)
c3 = df.Race_Starts.corr(df.Podiums)
c4 = df.Race_Starts.corr(df.Fastest_Laps)
c5 = df.Race_Starts.corr(df.Pole_Positions)

print('Correlations:')
print(f'    Race_Starts vs Championships: {c1}')
print(f'    Race_Starts vs Race_Wins: {c2}')
print(f'    Race_Starts vs Podiums: {c3}')
print(f'    Race_Starts vs Fastest_Laps: {c4}')
print(f'    Race_Starts vs Pole_Position: {c5}')
```
