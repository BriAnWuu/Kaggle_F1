```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('F1DriversDataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# countries represented
df.Nationality.value_counts().head(10)
```

```python
# number of champions
df[df.Champion == True].shape[0]
```

```python
# driver count by number of championships awarded
df[df.Champion == True].Championships.value_counts()
```

```python
plt.hist(df[df.Champion == True].Championships, bins = range(1, 9), align = 'left', rwidth = 0.9)
plt.xlabel('Championships')
plt.ylabel('Drivers')
plt.show()
```

```python
# drivers and active years
_, _, counts = plt.hist(df.Years_Active, bins = range(df.Years_Active.min(), df.Years_Active.max() + 2), align = 'left', rwidth = 0.9)
plt.xticks(range(df.Years_Active.min(), df.Years_Active.max() + 1))
plt.xlabel('Active Years')
plt.ylabel('Drivers')
plt.bar_label(counts)
plt.show()
```

```python
# visualize with heat map
corr_df = df[['Championships', 'Race_Starts', 'Pole_Positions', 'Race_Wins', 'Podiums', 'Fastest_Laps', 'Points', 'Years_Active']]
corr_df.head()
```

```python
plt.figure(figsize = (15, 8))
mask = np.triu(np.ones_like(corr_df.corr(), dtype = bool))
heatmap = sns.heatmap(corr_df.corr(), cmap = 'crest', mask = mask, annot = True, linewidths = .5)
heatmap.set_title('Correlation Heatmap', fontdict = {'fontsize' : 15}, pad = 12)
plt.show()
```
